```python
import random as rd
import string
from enum import Enum 
```

```python
class Pessoa:
    def __init__(self): 
        self.set_idade(0)
        self.set_nome('Indefinido')
        self.set_sexo('I')
    
    def __str__(self): 
        string_resultante = "Pessoa [Nome: " + self.get_nome() + \
        ", Idade: " + str(self.get_idade()) + \
        ", Gênero: " + self.get_sexo() + "]"
        
        return string_resultante 

    def set_nome(self, nome):
        self.__nome = nome   

    def set_idade(self, idade):
        self.__idade = idade

    def set_sexo(self, sexo):
        if (sexo == 'I' or sexo == 'i'):
            self.__sexo = 'I'
        elif (sexo == 'M' or sexo == 'm'):
            self.__sexo = 'M'
        elif (sexo == 'F' or sexo == 'f'):
            self.__sexo = 'F'

    def get_nome(self):
        return self.__nome

    def get_idade(self):
        return self.__idade

    def get_sexo(self):
        return self.__sexo
```

```python
class NivelAlunos(Enum):
    INDEFINIDO = 0
    FUNDAMENTAL = 1
    MEDIO = 2
    GRADUACAO = 3
    MESTRADO = 4
    DOUTORADO = 5
    
    def __str__(self):
        return self.name
    
class Aluno(Pessoa):
    def __init__(self, nome='Indefinido', sexo='I', idade=0, matricula='Indefinida', 
                 nivel=NivelAlunos.INDEFINIDO):        
        self.set_nome(nome)
        self.set_idade(idade)
        self.set_sexo(sexo)
        self.matricula = matricula 
        self.nivel = nivel         
        
    def __str__(self):
        string_resultante = "Aluno [Nome: " + self.get_nome() + \
        ", Idade: " + str(self.get_idade()) + \
        ", Gênero: " + self.get_sexo() + \
        ", Matrícula: " + self.matricula + \
        ", Nível: " + str(self.nivel) + "]"
        
        return string_resultante 
                   
    @property                    
    def matricula(self):         
        return self.__matricula   

    @matricula.setter
    def matricula(self, matricula):   
        self.__matricula = matricula       
    
    @property
    def nivel(self):
        return self.__nivel
       
    @nivel.setter
    def nivel(self, nivel):
        self.__nivel = nivel               
```

```python
class Categoria(Enum):
    INDEFINIDA = 0
    PROGRAMADOR = 1
    ANALISTA = 2
    GERENTE = 3
    def __str__(self):
        return self.name
    
class Funcionario(Pessoa):
    def __init__(self, nome='Indefinido', sexo='I', idade=0, ingresso=0, 
                 categoria=Categoria.INDEFINIDA):        
        self.set_nome(nome)
        self.set_idade(idade)
        self.set_sexo(sexo)
        self.ingresso = ingresso
        self.categoria = categoria    
    
    def __str__(self):
        string_resultante = "Funcionario [Nome: " + self.get_nome() + \
        ", Idade: " + str(self.get_idade()) + \
        ", Gênero: " + self.get_sexo() + \
        ", Ingresso: " + str(self.ingresso) + \
        ", Categoria: " + str(self.categoria) + "]"
        
        return string_resultante 
                   
    @property                    
    def categoria(self):         
        return self.__categoria

    @categoria.setter
    def categoria(self, categoria):   
        self.__categoria = categoria
    
    @property
    def ingresso(self):
        return self.__ingresso
       
    @ingresso.setter
    def ingresso(self, ingresso):
        if ingresso > 2020:
            raise ValueError("Ano de ingresso deve ser menor ou igual ao ano atual.")
        else:
            self.__ingresso = ingresso                 
```

```python
people = []

for _ in range(10):
    choice = rd.randint(0, 2)
    nome = ''.join(rd.choices(string.ascii_uppercase + string.digits, k=10))
    idade = rd.randint(0, 100)
    sexo = ('F', 'I', 'M')[rd.randint(0,2)]
    p = None
    if choice == 0:
        p = Pessoa()
        p.set_idade(idade)
        p.set_nome(nome)
        p.set_sexo(sexo)
    elif choice == 1:
        matricula = str(rd.randint(0, 100000))
        nivel = (NivelAlunos.INDEFINIDO, NivelAlunos.FUNDAMENTAL, NivelAlunos.MEDIO, NivelAlunos.GRADUACAO,\
                 NivelAlunos.MESTRADO, NivelAlunos.DOUTORADO)[rd.randint(0,5)]
        p = Aluno(nome, sexo, idade, matricula, nivel)
    elif choice == 2:
        ingresso = rd.randint(0, 2020)
        categoria = (Categoria.INDEFINIDA, Categoria.PROGRAMADOR, Categoria.ANALISTA, Categoria.GERENTE)[rd.randint(0,3)]
        p = Funcionario(nome, sexo, idade, ingresso, categoria)
    else:
        print("Something went wrong :(")
        exit(1)
    people.append(p)
```

```python
for p in people:
    print(p)
```
